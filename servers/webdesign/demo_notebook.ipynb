# WebDesign MCP Demo - Interactive Tutorial

This notebook demonstrates how to use the WebDesign MCP server to generate React websites.

## What You'll Learn

- How to analyze a web project use case
- Generate design specifications automatically
- Scaffold a complete React project
- Get pre-built component templates

## The Workflow

1. **Describe** your project → AI analyzes
2. **Review** design specs → Approve or iterate
3. **Generate** project files → Complete scaffold
4. **Develop** with components → Ready-to-use code

## Step 1: Set Up Environment

Run these commands in your terminal:

```bash
cd mcp_project
uv init
uv venv
source .venv/bin/activate
uv add mcp
```

## Step 2: Example Use Case - Solar Panel Company

Let's analyze a project for a solar panel installation company.

**Project Details:**
- Name: SolarPro
- Description: Landing page for residential solar panel installation
- Target: Homeowners in California (35-65 years)
- Goal: Generate quote requests
- Features: Hero, Features, Testimonials, Contact Form

## Step 3: Expected Output

The MCP server will analyze and return:

```json
{
  "template": {
    "name": "Landing Page",
    "type": "landing_page"
  },
  "palette": {
    "name": "Energy",
    "primary": "#059669",
    "vibe": "Sustainable, eco-friendly"
  },
  "stack": {
    "name": "Modern React",
    "technologies": {
      "framework": "React",
      "build": "Vite",
      "styling": "Tailwind CSS"
    }
  },
  "components": [
    "Hero",
    "FeatureCard",
    "Testimonial",
    "ContactForm"
  ]
}
```

## Step 4: Generated Project Structure

After running `generate_project_scaffold`:

```
solarpro-landing/
├── specs/
│   └── design-specs.md          # Complete design documentation
├── src/
│   ├── components/
│   │   └── ui/                  # Button, Card, Input
│   ├── pages/                   # Page components
│   └── App.tsx                  # Main app
├── package.json                 # Dependencies
├── tailwind.config.js           # Your color palette
└── README.md                    # Project docs
```

## Step 5: Start Development

```bash
cd solarpro-landing
npm install
npm run dev
```

Open http://localhost:5173 to see your site!

## Available Tools

| Tool | Purpose |
|------|---------|
| `analyze_use_case` | Generate design specs from description |
| `list_design_templates` | View website templates |
| `list_color_palettes` | Browse color schemes |
| `generate_project_scaffold` | Create project files |
| `get_component_template` | Get React component code |
| `list_available_components` | View component options |
| `get_design_tokens` | Get CSS variables |

## Design System

### Color Palettes

- **corporate** (#1E40AF) - Professional B2B
- **energy** (#059669) - Sustainable, eco-friendly
- **dark_modern** (#6366F1) - Tech, SaaS
- **minimal** (#18181B) - Luxury, portfolios

### Typography

- Font: Inter
- Scale: H1 3.5rem → Body 1rem
- Weights: 400, 500, 600, 700

### Tech Stack

- React 18 + TypeScript
- Vite (build tool)
- Tailwind CSS
- React Router
- Framer Motion
- Lucide Icons
